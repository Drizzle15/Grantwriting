```python
questions_array=["Does the issue affect other areas of persons life?","Are the other areas of the persons life affected by these issues?","Are the work activities of the people affected daily?","Is the target market global?","Is there health and social benefits with your product?","Does the health and social benifits by this product offered?","Is there any health or social beifits offered by this product?","Is there any change in the existing solution?","Has the exisiting solution changed much?","Are there physical impacts on the persons well-being?","Are there psychological impacts on the persons well-being?", "Is the well-being of the people impacted psychologically?","Are the activities of the target market affected daily?","Will this project unify different people?","Does this project unifies different people?", "Are the different people with this project unified?","Are the different people unified with this project?","Does the people from different background unified through this project?"]
```

```python
ammended_quest_array=[]
for question in questions_array:
  print(question)
  words = nltk.word_tokenize(question)
  print(words)
  #Tag each word with corresponding Part of Speech
  tagged = nltk.pos_tag(words)
  print('POS tagged words= ',tagged)
  
  word_str=""
  for word, tag in tagged:
    word_tone=""
    if tag not in('VBP','WP','.','WDT','WP$','WRB','VBZ','NNP'):
      #print(word)
      if tag in ('VBD','JJ','VB','JJR','VBN'):
        word= "is not" + " " + word
      word_str+=" "+  word
  ammended_quest_array.append(word_str)

print("\n\n")




print(ammended_quest_array)
```

```python
combined_array=[]

total=len(ammended_quest_array)

#for i in ammended_quest_array:
#  print(i)
"""
i=0
while(i!=total):
  combined_array.append(answer_array[i] + " " + "is" + " " + ammended_quest_array[i])
  i=i+1
"""

i=0
while(i!=total):
  combined_array.append( " " + ammended_quest_array[i]+".")
  i=i+1

print(combined_array)
```

```python
from gramformer import Gramformer
import torch

def set_seed(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

set_seed(1212)


gf = Gramformer(models = 1, use_gpu=False) # 1=corrector, 2=detector
```

```python
influent_sentences = combined_array
```

```python
for influent_sentence in influent_sentences:
    corrected_sentences = gf.correct(influent_sentence, max_candidates=1)
    print("[Input] ", influent_sentence)
    for corrected_sentence in corrected_sentences:
      print("[Correction] ",corrected_sentence[0])
    print("-" *100)
```
