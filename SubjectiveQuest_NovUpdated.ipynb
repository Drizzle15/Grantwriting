```python
from parrot import Parrot
import torch
import warnings
warnings.filterwarnings("ignore")
```

```python
def random_state(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

random_state(1234)
```

```python
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=False)
```

```python
#question="Does this issue prevent the affected from doing anything?"
question="What is the potential of this team?"
```

```python
answer = input(question)
```

```python
words = nltk.word_tokenize(question)
print(words)
  #Tag each word with corresponding Part of Speech
tagged = nltk.pos_tag(words)
print('POS tagged words= ',tagged)
word_str=""
for word, tag in tagged:
  if tag not in('WP','.','WDT','WP$','WRB','VBZ','VBP'):
      #print(word)
    word_str+=" "+  word
ammended_quest=word_str
```

```python
phrase =ammended_quest + " is" + " " + answer
```

```python
print(phrase)
```

```python

print("Input_phrase: ", phrase)
para_phrase = parrot.augment(input_phrase=phrase)
length=len(para_phrase)
print(para_phrase[length-1])
```
